```python
from turtle import Turtle, screen

screen = screen()
screen.bgcolor("black")
screen.setup(width=800, height=600)
screen.title("pong")

paddle = Turtle()
paddle.shape("square")
paddle.color("white")
paddle.shapesizes(stretch_wid=5, stretch_len=1)
paddle.penup()
paddle.goto(350, 0)

screen.exitonclick()
```
