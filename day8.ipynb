I continue with learning how Python assigns arguments to parameters:
Code example:

```python
def greet_with(name, location):
    print(f"Hello {name}")
    print(f"What is it like in {location}?")

greet_with("Angela", "London")  
greet_with(location="London", name="Angela")
```

coding exercise of a love calculator completion

```python
def calculate_love_score(name1, name2):
    # Combine both names and convert to lowercase
    combined_names = (name1 + name2).lower()

    # Count letters in "TRUE"
    true_count = 0
    for letter in "true":
        true_count += combined_names.count(letter)

    # Count letters in "LOVE"
    love_count = 0
    for letter in "love":
        love_count += combined_names.count(letter)

    # Combine the two counts into a two-digit number (as a string), then convert to int
    love_score = int(str(true_count) + str(love_count))

    # Print the love score
    print(love_score)

# Call your function with hard-coded values
calculate_love_score("Kanye West", "Kim Kardashian")
```
